# Task 4 : Exploratory Data Analysis

```python
# Feature Engineering

# 1. Converting values of senior citizen to yes or no
temp['SeniorCitizen'] = temp['SeniorCitizen'].apply(lambda x : 'yes' if x == 1 else 'no')
```

# Churn Counts

```python
churn_count = temp.groupby('Churn').agg(total_cust = ('Churn',"count")).reset_index()
fig, ax = plt.subplots(1, 2, figsize = (15,5))

bars = ax[0].bar(churn_count['Churn'], churn_count['total_cust'])
ax[0].set_title('Number of Churn vs Non-Churn Customers', fontsize = 12, fontweight = 'bold')
ax[0].set_xlabel('Churn')
ax[0].set_ylabel('Number of Customers')
ax[0].bar_label(bars)


ax[1].pie(churn_count['total_cust'], labels = churn_count['Churn'], autopct = '%0.1f%%')
ax[1].set_title('Churn Percent', fontsize = 12, fontweight = 'bold')

plt.tight_layout()
plt.show()
```

* *Number of **Churn** customers = **1869**.*
* *Number of **non-Churn** customers = **5174**.*
* ***Percentage of Churn customers = 26.5%**.*
* ***Percentage of Non-Churn customers = 74.5%**.*

```python
fig, ax = plt.subplots(1, 4, figsize = (15,5))

columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']

ax = ax.flatten()  # turns the 3x3 grid into a flat array of 9 Axes

for i, cols in enumerate(columns): 
    sns.countplot(data = temp, x = cols, hue = 'Churn', ax = ax[i])
    ax[i].set_title(f'Churn by {cols}', fontsize = 12, fontweight = 'bold')
    ax[i].set_xlabel(cols)
    ax[i].set_ylabel('Count')
    
    for container in ax[i].containers:
        ax[i].bar_label(container)          

plt.tight_layout()
plt.show()
```

* ***Male Churn = 930***
* ***Female Churn = 939***

```python
fig, ax = plt.subplots(3, 3, figsize = (15,12))

services = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

ax = ax.flatten()  # turns the 3x3 grid into a flat array of 9 Axes

for i, ser in enumerate(services): 
    sns.countplot(data = temp, x = ser, hue = 'Churn', ax = ax[i])
    ax[i].set_title(f'Churn by {ser}', fontsize = 12, fontweight = 'bold')
    ax[i].set_xlabel(ser)
    ax[i].set_ylabel('Count')
    
    for container in ax[i].containers:
        ax[i].bar_label(container)          

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.histplot(data = temp, x = 'tenure', bins = 73, hue = 'Churn')
plt.title('Churn by tenure', fontsize = 12, fontweight = 'bold')
plt.xlabel('Tenure(months)')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,5))
ax = sns.countplot(data = temp, x = 'PaymentMethod', hue = 'Churn')
plt.title('Churn by Payment Methods', fontsize = 12, fontweight = 'bold')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.tight_layout()
plt.show()
```
